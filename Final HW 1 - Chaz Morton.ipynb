### Handle missing values & label encoding:

```python
# Drop rows with NaN value for features 'RainToday' and 'RainTomorrow'

data.dropna(subset=['RainToday'], inplace=True)
data.dropna(subset=['RainTomorrow'], inplace=True)
```

We do not want to fill in these NaN for the target feature. Instead, we will do it for the other features

```python
# FOR CATEGORICAL FEATURES WITH NaN

# Get rows with no NaN values for features 'WindGustDir', 'WindDir9am', etc.

notnullWindGustDir = data.loc[data['WindGustDir'].notnull()]
notnullWindDir9am = data.loc[data['WindDir9am'].notnull()]
notnullWindDir3pm = data.loc[data['WindDir3pm'].notnull()]
```

```python
## LABEL ENCODING: CATEGORICAL FEATURES. KEEP NaN

from sklearn import preprocessing

# Label Encoder

label_encoder = preprocessing.LabelEncoder()

# Encode categorical features and for values that are not NaN

data["Location"] = label_encoder.fit_transform(data["Location"])
notnullWindGustDir['WindGustDir'] = label_encoder.fit_transform(notnullWindGustDir["WindGustDir"])
notnullWindDir9am['WindDir9am'] = label_encoder.fit_transform(notnullWindDir9am["WindDir9am"])
notnullWindDir3pm['WindDir3pm'] = label_encoder.fit_transform(notnullWindDir3pm["WindDir3pm"])
data['RainToday'] = label_encoder.fit_transform(data["RainToday"])
data['RainTomorrow'] = label_encoder.fit_transform(data["RainTomorrow"])
```

```python
# Update categorical features, keep NaN

data['WindGustDir'].update(notnullWindGustDir['WindGustDir'])
data['WindDir9am'].update(notnullWindDir9am['WindDir9am'])
data['WindDir3pm'].update(notnullWindDir3pm['WindDir3pm'])
```

We are keeping NaN because we want to encode the categorical values, then replace the NaN using those encoded values. To do this we will randomly sample from a gaussian/normal distribution of these label encoded categorical values.

```python
# Check for missing values

missing_values = data.isnull().sum()
features_with_missing = missing_values[missing_values > 0].index
features_with_missing
```

```python
# VISUALIZE NORM. DIST. OF FEATURE 'Sunshine'

import scipy.stats as stats

feature = "Sunshine"

# Calculating mean and standard deviation 
d = data[feature].loc[data[feature].notnull()]
mean = data[feature].mean()
sd = data[feature].std()

print('Normal distribution for ', feature, ' feature:')
plt.plot(d.sort_values(), stats.norm.pdf(d.sort_values(), mean, sd)) 
plt.show() 
```

```python
# FOR CATEGORICAL FEATURES ONLY

# Import library
from scipy.stats import norm

# Get random values from distribution. Used to fill NaN values

def impute_with_gaussian_categorical(feature):
    mean = data[feature].mean()
    std_dev = data[feature].std()
    size = data[feature].isnull().sum()
    
    # Generate random values from gaussian distribution
    random_values = np.random.normal(mean, std_dev, size)
    for i, value in enumerate(random_values):
        if value < 0:
            random_values[i] = 0
        random_values[i] = round(random_values[i])
    print("Random values for feature: ", feature, "\n", random_values)
    return random_values


categorical_features_with_missing = ['WindGustDir', "WindDir9am", "WindDir3pm"]

# Impute missing values for each feature
for feature in categorical_features_with_missing:
    data.loc[data[feature].isnull(), feature] = impute_with_gaussian_categorical(feature)
```

Here, we fill in the NaN values for the categorical features that still had NaN. 

```python
# FOR NUMERICAL FEATURES ONLY

# Get random values from distribution. Used to fill NaN values

def impute_with_gaussian_numerical(feature):
    mean = data[feature].mean()
    std_dev = data[feature].std()
    size = data[feature].isnull().sum()
    
    # Generate random values from gaussian distribution
    random_values = np.random.normal(mean, std_dev, size)
    for i, value in enumerate(random_values):
        random_values[i] = round(value)
    print("Random values for feature: ", feature, "\n", random_values)
    return random_values


numerical_features_with_missing = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 
                                     'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 
                                     'Humidity3pm','Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 
                                     'Temp9am', 'Temp3pm']

# Impute missing values for each feature
for feature in numerical_features_with_missing:
    data.loc[data[feature].isnull(), feature] = impute_with_gaussian_numerical(feature)
```

```python
# Some features it doesn't make sense to have values below 0. Let's fix this: 

set_min_zero = ['Rainfall', 'Evaporation', 'Sunshine', 
                'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 
                'Humidity3pm', 'Cloud9am', 'Cloud3pm']

for feature in set_min_zero:
    data[feature].mask(data[feature] < 0, 0, inplace=True)
```

We can now see that the count for all features are the same. Meaning we have handled the missing values from the dataset, and are left with 140,787 rows. 

### Handle outliers:

```python
# Show a boxplot with outliers

sns.boxplot(data['MinTemp'])
plt.title('Box Plot for Outlier Detection')
plt.show()
```

```python
# Get rid of the outliers by using the IQR for reach feature

outliers = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm']

for feature in outliers:
    # Calculate IQR
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1

    # Threshold
    threshold = 1.5

    # Remove outliers
    data = data[(data[feature] > (Q1 - threshold * IQR)) & (data[feature] < (Q3 + threshold * IQR))]
```

### Feature Selection:

```python
# Split data into X and y

X = data.drop(['Date', 'RainTomorrow'], axis=1)  # Features excluding the target feature
y = data['RainTomorrow']  # Target feature
```

# 1 continued... Split Data:

### Split the data into training and testing sets:

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Display the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# 3 & 4 & 5: Model Building, Model Training and Validation, Model Evaluation

- Use SKLearn or PyTorch to build a neural network model. Make it a simple multilayer fully connected neural network. At least one hidden layer required.
- Experiment with different architectures (number of layers and neurons).
- Choose an appropriate activation function for the hidden layers and the output layer.

```python
# THE MODEL

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


# Create MLPClassifier with activation ReLU, and one hidden layer of 64 neurons

model = MLPClassifier(hidden_layer_sizes=(64), activation='relu')

# Train the model

model.fit(X_train, y_train)

# Predictions on the testing set

y_pred = model.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(conf_matrix)
```

### Find the best hyperparameters: 

```python
learning_rates = [0.001, 0.01, 0.1]
batch_sizes = [16, 32, 64]

best_accuracy = 0
best_hyperparameters = {}

for lr in learning_rates:
    for batch_size in batch_sizes:
        
        model = MLPClassifier(hidden_layer_sizes=(64), activation='relu', learning_rate_init=lr, batch_size=batch_size)
        model.fit(X_train, y_train)
        y_val_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_val_pred)

        # Update best hyperparameters
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            print("Best accuracy: ", best_accuracy)
            best_hyperparameters = {'learning_rate': lr, 'batch_size': batch_size}

print("Best Hyperparameters:", best_hyperparameters)
```

This takes a little bit. But once we find out best hyperparameters for learning rates and batch sizes, we can specify them and train our model again.

```python
model = MLPClassifier(hidden_layer_sizes=(64), activation='relu', learning_rate_init=0.001, batch_size=64)

# Train the model

model.fit(X_train, y_train)

# Predictions on the testing set

y_pred = model.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(conf_matrix)
```

- Confusion Matrix:
    - We start with our true positives (TP) in the top left. It is the amount of times that model correctly predicts the positive class. The positive class is that it does rain tomorrow ('RainTomorrow' == 1).
    - False positives (FP) on the top right is the instances where the model predicts the positive class, but the true class is negative (Type I error).
    - True negative (TN) on bottom right is how many times it correctly predicts the negative class ('RainTomorrow' == 0).
    - False negative (FN) is where the model predicts the negative class, but the true class is positive (Type II error).

The model does well in accuracy, but it struggles in other metrics. It is best to aim for a well rounded model where all of these metrics are relatively high. 

- Accuracy measures the overall correctness of the model.
- Precision focuses on the accuracy of positive predictions.
- Recall measures the ability to capture the relevant instances of the positive class.
- F-1 score is like a mean of the precision and the recall.
